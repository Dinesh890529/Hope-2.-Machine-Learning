```python
import pandas as pd
```

```python
dataset = pd.read_csv("50_Startups.csv")
dataset
```

```python
dataset = pd.get_dummies(dataset, dtype = int, drop_first=True)
dataset
```

```python
dataset.info()
```

```python
dataset.columns
```

```python
independent = dataset[['R&D Spend', 'Administration', 'Marketing Spend', 'State_Florida', 'State_New York']]
```

```python
dependent = dataset[['Profit']]
```

```python
from sklearn.preprocessing import StandardScaler
standardize = StandardScaler()
scaled_independent = standardize.fit_transform(independent)
scaled_dependent = standardize.fit_transform(dependent)
```

```python
scaled_independent
```

```python
scaled_dependent
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(scaled_independent,scaled_dependent,test_size=0.3,random_state=0)
```

```python
from sklearn.svm import SVR
regressor = SVR(kernel="linear")
regressor.fit(X_train,y_train)
```

```python
regressor.coef_
```

```python
regressor.intercept_
```

```python
regressor.n_support_
```

```python
regressor.support_
```

```python
y_pred = regressor.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r_squared = r2_score(y_pred,y_test)
```

```python
r_squared
```

```python
import pickle
filename = "FinalModel_SVR_linear.sav"
pickle.dump(regressor,open(filename,'wb'))
```
