```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("50_Startups.csv")
```

```python
dataset
```

```python
dataset.shape
```

```python
dataset = pd.get_dummies(dataset,drop_first=True)
```

```python
dataset
```

```python
dataset.shape
```

```python
dataset.columns
```

```python
Independent = dataset[['R&D Spend', 'Administration', 'Marketing Spend', 'State_Florida', 'State_New York']]
```

```python
dependent = dataset[['Profit']]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(Independent,dependent,test_size=0.30, random_state=0)
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor(criterion = 'squared_error', splitter = 'best')
regressor1 = DecisionTreeRegressor(criterion = 'squared_error', splitter = 'random')
regressor = regressor.fit(X_train,y_train)
regressor1 = regressor1.fit(X_train,y_train)

regressor2 = DecisionTreeRegressor(criterion = 'friedman_mse', splitter = 'best')
regressor3 = DecisionTreeRegressor(criterion = 'friedman_mse', splitter = 'random')
regressor2 = regressor2.fit(X_train,y_train)
regressor3 = regressor3.fit(X_train,y_train)

regressor4 = DecisionTreeRegressor(criterion = 'absolute_error', splitter = 'best')
regressor5 = DecisionTreeRegressor(criterion = 'absolute_error', splitter = 'random')
regressor4 = regressor4.fit(X_train,y_train)
regressor5 = regressor5.fit(X_train,y_train)

regressor6 = DecisionTreeRegressor(criterion = 'poisson', splitter = 'best')
regressor7= DecisionTreeRegressor(criterion = 'poisson', splitter = 'random')
regressor6 = regressor6.fit(X_train,y_train)
regressor7 = regressor7.fit(X_train,y_train)
```

```python
from sklearn import tree
tree.plot_tree(regressor)
plt.show()
```

```python
y_pred = regressor.predict(X_test)
y_pred1 = regressor1.predict(X_test)
y_pred2 = regressor2.predict(X_test)
y_pred3 = regressor3.predict(X_test)
y_pred4 = regressor4.predict(X_test)
y_pred5 = regressor5.predict(X_test)
y_pred6 = regressor6.predict(X_test)
y_pred7 = regressor7.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r_squared_score = r2_score(y_test,y_pred)
r_squared_score1 = r2_score(y_test,y_pred1)
r_squared_score2 = r2_score(y_test,y_pred2)
r_squared_score3 = r2_score(y_test,y_pred3)
r_squared_score4 = r2_score(y_test,y_pred4)
r_squared_score5 = r2_score(y_test,y_pred5)
r_squared_score6 = r2_score(y_test,y_pred6)
r_squared_score7 = r2_score(y_test,y_pred7)
```

```python
print(r_squared_score,r_squared_score1,r_squared_score2,r_squared_score3,r_squared_score4,r_squared_score5,r_squared_score6,r_squared_score7)
```

# Conclusion:



### Decision tree regressor with default parameter of " criterion = 'absolute_error', splitter = 'best' " gives the best r_squared_score = 0.94
