# Regression Assignment:

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv("insurance_prediction_dataset.csv")
```

```python
dataset
```

```python
dataset.shape
```

```python
dataset.info()
```

```python
dataset = pd.get_dummies(dataset, dtype=int, drop_first= True)
```

```python
dataset
```

```python
dataset.columns
```

```python
independent = dataset[['age', 'bmi', 'children', 'sex_male', 'smoker_yes']]
```

```python
dependent = dataset[['charges']]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(independent, dependent, test_size=0.3, random_state=0)
```

## MultiLinear Regression:

```python
from sklearn.linear_model import LinearRegression
regressor_MultiLinear = LinearRegression()
regressor_MultiLinear.fit(X_train,y_train)
```

```python
regressor_MultiLinear.coef_
```

```python
regressor_MultiLinear.intercept_
```

```python
y_pred_MultiLinear = regressor_MultiLinear.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r_squared_MultiLinear = r2_score(y_pred_MultiLinear,y_test)
```

```python
r_squared_MultiLinear
```

## Support Vector Machine Regression:

```python
from sklearn.svm import SVR
regressor_SVR = SVR(kernel="linear")
regressor_SVR.fit(X_train,y_train)
```

```python
y_pred_SVR = regressor_SVR.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r_squared_SVR = r2_score(y_pred_SVR,y_test)
```

```python
r_squared_SVR
```
