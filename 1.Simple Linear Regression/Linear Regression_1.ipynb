# Linear Regression Algorithm:

### Importing Library pandas

```python
import pandas as pd
```

### Load the salary dataset:

```python
dataset = pd.read_csv('D:\HOPE\AI Course Tamil\Machine Learning\Salary_Data.csv')
```

```python
dataset
```

### Creating Independent Variable:

```python
independent = dataset[['YearsExperience']]
independent
```

### Creating Dependent Variable:

```python
dependent = dataset[['Salary']]
dependent
```

### Training and Test data splits:

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(independent, dependent, test_size = 0.30, random_state = 0)
```

### Importing Linear Algorithm and fitting Training data suing SKlearn Library:

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train,y_train)
```

### To get the value of Slope or weight (m):

```python
weight_or_slope = regressor.coef_
weight_or_slope
```

### To get the value of Bias or Intercept (c) of equation y=mx+c:

```python
bias_or_intercept = regressor.intercept_
bias_or_intercept
```

### Prediction:

```python
y_pred = regressor.predict(X_test)
```

### Validation Parameter R-Squared:

```python
from sklearn.metrics import r2_score
r_score = r2_score(y_test,y_pred)
```

### R-Squared score:

```python
r_score
```

### Using Pickle Library to load and save our model for future use:

```python
import pickle
filename = "finalized_model_linears.sav"
```

```python
pickle.dump(regressor,open(filename, 'wb'))
```

```python
loaded_model = pickle.load(open("finalized_model_linears.sav", 'rb'))
result = loaded_model.predict([[15]])
```
